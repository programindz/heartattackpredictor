```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df = pd.read_csv("heart.csv", header = 0)
# reading the csv data and converting it into a data frame
# header= 0 means 0th row of csv represents headings
```

```python
df = df.astype(float) 
#changing the datatype of each column to float64
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn import svm
```

```python
X = np.array(df.drop(['output'],axis = 'columns'))
# removing the output column from dataframe(output is dropped because it is to be predicted)
# try to remove/drop the columns that have less impact on output 

y= np.array(df['output'])
# y represents column to be predicted

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 15)
#spliting the data into training(80%) and testing sets(20%)

clf = make_pipeline(StandardScaler(),svm.SVC(kernel = 'linear'))
# clf is our classifier/model
# StandardScaler is used to scale the X data against the y data
# pipeline integrates the use of scaling along with model(SVC in our case)
# kernel parameter uses kernel functions to transform nonlinear training data to linear

clf.fit(X_train, y_train)
#this is actual training of data

clf.score(X_test, y_test)
#accuracy of our model
```

```python
predict_data = [30.0, 0.0, 3.0, 123.0, 154.0, 1.0, 2.0, 115.0, 0.0, 1.1]
#data to be predicted, you can provide custom data here

result = clf.predict(np.array(predict_data).reshape(1, -1))
#prediction by our model(result will be either 0:low chances of cancer or 1: high chance of cancer)
print(result)
if result == [1]:
    print("Person has more chances of having a Heart Attack")
else:
    print("Person has less chances of having a Heart Attack")
```

```python
from sklearn.linear_model import LogisticRegression
regr = make_pipeline(StandardScaler(), LogisticRegression(random_state=0))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.2)
regr.fit(X_train, y_train)
regr.score(X_test, y_test)
```

```python
res = regr.predict(np.array(predict_data).reshape(1, -1))
print(res)
if res == [1]:
    print("Person has more chances of having a Heart Attack")
else:
    print("Person has less chances of having a Heart Attack")
```
